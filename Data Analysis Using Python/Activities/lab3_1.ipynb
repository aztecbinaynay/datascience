```python
# load the diabetes.csv file
# import numpy and pandas

import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')
filepath = '/content/drive/MyDrive/Colab Notebooks/diabetes.csv'
data = pd.read_csv(filepath)
```

```python
# Display the last 20 records
data.tail(20)
```

```python
# Change the Outcome column to Diagnosis
data.rename(columns ={'Outcome' : 'Diagnosis'}, inplace = True)
data
```

```python
# Create a new column Classification that display "Diabetes" if the value of Diagnosis is 1 , 
# otherwise "No Diabetes"

data['Classification'] = np.where(data['Diagnosis'] == 1, 'Diabetes', 'No Diabetes')
data
```

```python
# Create a new dataframe "withDiabetes" that gathers data with diabetes

withDiabetes = data[(data['Diagnosis'] ==  1)]
withDiabetes
```

```python
# Create a new dataframe "Pedia" that gathers data with age 0 to 19

Pedia = data[(data['Age'] >=  0) & (data['Age'] <=  19)]
Pedia

#! a better way of doing this is by using the code:
#? data[np.logical_and(data['Age'] >= 0, data["Age"] <= 19)]
```

```python
# Create a new dataframe "Adult" that gathers data with age greater than 19

Adult = data[(data['Age'] >  19)]
Adult
```

```python
# Use numpy to get the average age and glucose value.
print('the average age in the data is', data.Age.mean())
print('the average glucose in the data is', data.Glucose.mean())
```

```python
# Use numpy to get the median age and glucose value.

print('the median age in the data is', data.Age.median())
print('the median glucose in the data is', data.Glucose.median())
```

```python
# Use numpy to get the middle values of glucose and age.
# Since the median value of a set is the same as the middel value of that set, then 
# findin the median (as what we did in the previous exercise) is the correct solution

print('the middle value in column Age in the data is', data.Age.median())
print('the middle value in column glucose in the data is', data.Glucose.median())
```

```python
# Use numpy to get the standard deviation of the skinthickness.

print("Standard Deviation of skinthickness:", data.SkinThickness.std())
```
