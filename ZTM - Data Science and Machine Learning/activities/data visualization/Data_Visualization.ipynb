# Matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

# import the data
x = [1,2,3,4]
y = [11,22,33,44]

#setup the plot
fig, ax = plt.subplots(figsize = (10,10)) #(width, height)

#plot the x and y values
ax.plot(x,y)

#add labels to the plot
ax.set(title = "Simple Plot",
       xlabel = "x-axis",
       ylabel = "y-axis")

#save and show (save the whole figure)
fig.savefig("simple-plot.png")
```

```python
fig, ax = plt.subplots(figsize = (6,4))

# Plot
    # Plot histogram
avocado.plot(kind = "hist", density = True, alpha = 0.65, bins = 15) # change density to true, because KDE uses density
    # Plot KDE
avocado.plot(kind = "kde")

    # Quantile lines
quant_5, quant_25, quant_50, quant_75, quant_95 = avocado.quantile(0.05), avocado.quantile(0.25), avocado.quantile(0.5), avocado.quantile(0.75), avocado.quantile(0.95)
quants = [[quant_5, 0.6, 0.16], [quant_25, 0.8, 0.26], [quant_50, 1, 0.36],  [quant_75, 0.8, 0.46], [quant_95, 0.6, 0.56]]
for i in quants:
    ax.axvline(i[0], alpha = i[1], ymax = i[2], linestyle = ":")


# X
ax.set_xlabel("Average Price ($)")
    # Limit x range to 0-4
x_start, x_end = 0, 4
ax.set_xlim(x_start, x_end)

# Y
ax.set_ylim(0, 1)
ax.set_yticklabels([])
ax.set_ylabel("")

# Annotations
ax.text(quant_5-.1, 0.17, "5th", size = 10, alpha = 0.8)
ax.text(quant_25-.13, 0.27, "25th", size = 11, alpha = 0.85)
ax.text(quant_50-.13, 0.37, "50th", size = 12, alpha = 1)
ax.text(quant_75-.13, 0.47, "75th", size = 11, alpha = 0.85)
ax.text(quant_95-.25, 0.57, "95th Percentile", size = 10, alpha =.8)

# Overall
ax.grid(False)
ax.set_title("Avocado Prices in U.S. Markets", size = 17, pad = 10)

    # Remove ticks and spines
ax.tick_params(left = False, bottom = False)
for ax, spine in ax.spines.items():
    spine.set_visible(False)
    
plt.show()
```

# Two different ways to create subplots

```python
# option 1 for creating subplots

fig, ((ax1, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                            ncols = 2,
                                            figsize = (10,5))
# you can create multiple axes on a figure by using this line, ax1 and ax2 belong in the same row and ax3 and ax4 belong in the same row (2nd). you need to specify the nrows, ncols, and optioanlly the figsize arguments in the .subplots()
ax1.plot(x,x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nutbutter_prices.keys(), nutbutter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# option 2 for creating subplots

fig, ax = plt.subplots(nrows = 2,
                       ncols = 2,
                       figsize= (10,10)
                       )
#plot to each different index. Call the ax created in a given (row, column)
ax[0,0].plot(x,x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nutbutter_prices.keys(), nutbutter_prices.values())
ax[1,1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matlabplotlib OO version?)
* plotting something simple? Use the pyplot
* plotting something advanced? Use the OO method



```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.sort_values(by=["age"], inplace = True)
over_50.reset_index(inplace=True)
over_50
```

```python
#pandas plotting method (FLOP)
over_50.plot(kind = "scatter",x = "age", y = "chol", c = "target"); #c is the color
```

```python
# OO method (still using the pandas .plot() method)
fig, ax = plt.subplots(figsize = (10,6))
over_50.plot(kind = "scatter",
             x = "age",
             y = "chol",
             c = "target", 
             ax=ax); #c is the color

#change limitation of x
# ax.set_xlim([45,100])
```

```python
# OO method (w/o using the .plot() of pandas)

fig, ax = plt.subplots(figsize = (10, 5))

#plot the Data

scatter = ax.scatter(x = over_50['age'],
                     y = over_50["chol"],
                     c = over_50["target"]); #! add cmap = "" to change the color scheme
# customize th plot
ax.set(title = "Heart Disease and Cholesterol Levels",
       xlabel = "Age",
       ylabel = "cholesterol")

#set the legend
ax.legend(*scatter.legend_elements(), title = "Target");
#the c parameter in ax.scatter will be cheked and then the column of Target is queried to see what values are unique so that they can be put in the legend.

# add a horizontal line for the mean

ax.axhline(y = over_50.chol.mean(),
         ls = "--");
```

```python
plt.style.use("dark_background")

# Subplot of chol, age, and thalach
fig, (ax0, ax1) = plt.subplots(nrows = 2,
                               ncols = 1,
                               figsize  = (15,10),
                               sharex=True)

# add data to axis 0

scatter = ax0.scatter(x = over_50.age, y = over_50.chol, c= over_50.target, cmap="winter")
ax0.set(title =" Cholesterol vs Age",ylabel= "cholesterol")
ax0.legend(*scatter.legend_elements(), title = "Target")
ax0.axhline(y = over_50.chol.mean(), ls = "--", color = "red");
ax0.set_xlim([50, 80])
ax0.set_ylim([100,600])
# add data to axis 1

scatter1 = ax1.scatter(x = over_50.age, y = over_50.thalach, c= over_50.target, cmap="winter")
ax1.set(title =" Thalach vs Age", xlabel = "Age",ylabel= "thalach")
ax1.legend(*scatter.legend_elements(), title = "Target", )
ax1.axhline(y = over_50.thalach.mean(), ls = "--", color = "red");
ax1.set_xlim([50, 80])
ax1.set_ylim([50, 200])

# add a title to the fig
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
