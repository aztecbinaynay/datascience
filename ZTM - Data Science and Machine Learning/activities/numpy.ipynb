## Dot prodcut Example (nut butter sales)


```python
np.random.seed(0)
sales = np.random.randint(20, size=(5,3))
sales
```

```python
# create weekly sales data fram
weekly_sales = pd.DataFrame(sales, index = ["Mon", "Tue","Wed", "Thu","Fri"], columns = ["Almond Butter", "Peanut Butter","Cashew Butter"])
weekly_sales
```

```python
# creat the prices array
prices = np.array([10,8,12]) #! this array will be (3,) meaning 3 columns and 1 row, however this will not work in the data frame as intended thefore prices will be reshapped as (1,3) even though the structure of the 1-d array is the same

# create butter prices datafram
butter_prices = pd.DataFrame(prices.reshape(1,3), index = ["Price"], columns =["Almond Butter", "Peanut Butter","Cashew Butter"])
butter_prices
```

```python
sales.shape, prices.reshape(1,3).shape
```

```python
# since inner dimensions of the array are not the same we need to transpose prices (i.e. we need (a,b)x(b,c) or (a,1)x(b,c))
total_sales = sales.dot(prices.T)
total_sales
```

```python
butter_prices, weekly_sales
```

```python
total = pd.DataFrame(total_sales,
                    index = ["Mon", "Tue","Wed", "Thu","Fri"],columns = ["total"])
total
```

```python
total_weekly_sales = weekly_sales.join(total)
# or weekly_sales["Total"] = total

total_weekly_sales
```

## Sorting Arrays

```python
# code above keeps the same random numbers every time this cell is run
array10 = np.array([[2, 3, 8, 1, 3],
        [3, 3, 7, 0, 1],
        [9, 9, 0, 4, 7]])
array10
```

```python
np.argsort(array10, axis  =1)
# affects the elements per row when axis = 1 and columns when axis = 0. The resulting elements in the array show which index in the original array that element should to achieve sort. For example if the original array is [2, 3, 8, 1, 3] and the argsort equivalent is [3, 0, 1, 4, 2] then element in index 3 in the orignal array should be at the beginning of the array, index 0 element of the orig should be in the 2nd index... and so on....
```

```python
np.sort(array10)
#it seems this sorting function sorts the elements per row.
```

```python
np.argmin(array10), np.argmax(array10), np.argmax(array10, axis = 0)
# argmin and argmax checks which index in the array has the lowest and highest digit. when coupled with axis = 0 or 1 then the program gives an index of the elements that are the lowest/highest in each row or column. Here axis = 0 checks the columns
```
