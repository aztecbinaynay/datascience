## vacc_meta cleaning

```python
vacc_meta.head()
```

```python
vacc_meta.info()
```

```python
# extract ASEAN countries and replace the ISO3 name with country and replace ISO3 values with country name
asean = {"BRN":"Brunei", "MMR":"Myanmar", "KHM":"Cambodia", "TLS":"Timor-Leste", "IDN":"Indonesia", "LAO":"Lao People's Democratic Republic", "MYS":"Malaysia", "PHL":"Philippines", "SGP":"Singapore", "THA":"Thailand","VNM":"Viet Nam"}
```

```python
vacc_meta["ISO3"] = vacc_meta["ISO3"].replace(asean)
```

```python
vacc_meta["ISO3"]
```

```python
# create a datafram containing only ASEAN countries
vacc_meta = vacc_meta[vacc_meta["ISO3"].isin(asean.values())]
vacc_meta.rename(columns={"ISO3":"Country"}, inplace=True)
```

```python
vacc_meta.Country.value_counts()
```

```python
vacc_meta.info()
```

```python
# drop comment and end_date column as well as data_source
vacc_meta.drop(["COMMENT", "END_DATE", "DATA_SOURCE"], axis=1, inplace=True)
```

```python
vacc_meta.drop("PRODUCT_NAME", axis=1, inplace=True)
vacc_meta.sort_values(by=["Country"], inplace=True)
```

```python
vacc_meta.head()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
sns.countplot(y="VACCINE_NAME", data=vacc_meta)
ax.set_yticklabels(labels = vacc_meta["VACCINE_NAME"],horizontalalignment = "left")
ax.tick_params(axis='y', direction='in',pad=-10)
ax.set_title("Most Used Vaccines")
```
