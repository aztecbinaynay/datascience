```python
import pandas as pd
import folium 
from folium.plugins import StripePattern
import geopandas as gpd
import numpy as np
```

```python
sample_map = folium.Map(location=[48, -102], zoom_start=4)
sample_map
```

# Total Cases

```python
folium.Choropleth(
geo_data=df2,
data=df2,
columns=['country','cases_total'],
key_on="feature.properties.country",
fill_color='YlGnBu',
fill_opacity=1,
line_opacity=0.2,
legend_name="covid cases",
smooth_factor=0,
Highlight= True,
line_color = "#0000",
name = "Total Covid Cases",
show=False,
overlay=True,
nan_fill_color = "White"
).add_to(sample_map)

sample_map
```

# Total Deaths

```python
sample_map_deaths = folium.Map(location=[48, -102], zoom_start=4)
sample_map_deaths
```

```python
folium.Choropleth(
geo_data=df2,
data=df2,
columns=['country','cases_death'],
key_on="feature.properties.country",
fill_color='YlGnBu',
fill_opacity=1,
line_opacity=0.2,
legend_name="covid deaths",
smooth_factor=0,
Highlight= True,
line_color = "#0000",
name = "Total Covid deaths",
show=False,
overlay=True,
nan_fill_color = "White"
).add_to(sample_map_deaths)

sample_map_deaths
```
